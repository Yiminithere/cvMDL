```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as spo
from statsmodels.distributions.empirical_distribution import ECDF
from scipy.interpolate import interp1d

class MvECDF: 
    np.random.seed(123456790)
    '''
    This class is used for computation and evaluation of multidimensional empirical CDFs
    '''
    
    def __init__(self, data):
        np.random.seed(123456790)
        self._data = np.array(data)
        self._dimension = self._data.shape
        if len(self._dimension) > 2:
            raise TypeError("Incorrect input data dimension.") 
        elif len(self._dimension) == 1 or self._dimension[1] == 1:
            self._ecdf = ECDF(self._data.flatten())
        else:
            self._argsort = {i: np.argsort(self._data[:,i]) for i in range(self._dimension[1])}
            self._sorted = {i: self._data[:,i][self._argsort[i]] for i in range(self._dimension[1])}

    def random_match(self, x, y):  
        '''
        random_match(x, y) finds the number of elements in an increasing sequence y that 
        are less than or equal to x
        '''
        n = len(y)
        if x < y[0]:
            return 0
        if x >= y[-1]:
            return n
        m = np.random.randint(0,n-1)
        if x <= y[m]:
            return self.random_match(x, y[:(m+1)])
        elif x > y[m]:
            return m + 1 + self.random_match(x, y[(m+1):])
          
    def ecdf_eval(self, x):
        '''
        ecdf_eval(x) evaluates the empirical CDF of the input data at x
        '''
        if type(x) != list or type(x) != np.ndarray:
            x = [x]
        x = np.array(x).flatten()
        
        if len(self._dimension) == 1 or self._dimension[1] == 1:
            if len(x) == 1:
                return self._ecdf(x.flatten())[0]
            else:
                raise TypeError("Inconsistent input data dimension.")
        else: 
            if len(x) != self._dimension[1]:
                raise TypeError("Inconsistent input data dimension.")
            else:
                A = np.zeros((len(x),self._dimension[0]))
                for i in range(len(x)):
                    ind = self.random_match(x[i], self._sorted[i])
                    if ind == 0:
                        return 0
                    else:
                        A[i,:][self._argsort[i][:ind]] = 1
                return np.sum(np.prod(A, axis = 0))/self._dimension[0]

            
class Utility:
    np.random.seed(123456790)
    
    '''
    This class contains all the generic functions used to develop MFDistributionLearning
    '''
    def __init__(self):
        np.random.seed(123456790)
        self._low = -1e10
        self._up = 1e10
    
    def subsets(self, n):
        
        '''
        compute powerset of range(n)
        '''
        if n == 0:
            return [[]]
        else:
            x = self.subsets(n-1)
            return x + [y + [n-1] for y in x]
       
    def model_list(self, n, m):
        
        '''
        compute powerset of range(n) with size <= m
        '''
        x = self.subsets(n)
        res = [y for y in x if 0<len(y)<=m]
        return res 
  
    def cost_match(self, c):
        
        '''
        get cost of each subset of c
        '''
        c = np.array(c)
        len_c = len(c)
        subset_list = self.model_list(len_c, len_c)
        res = [np.sum(c[S]) for S in subset_list]
        return res
    
    def J0_eval(self, x):

        '''
        estimate the J0 function of the empirical CDF of x (F_x): int_{R} F_x(z)(1-F_x(z)) dz
        '''
        x.sort()
        n = len(x)
        integrand_val = [k/n*(1-k/n) for k in range(1,n)]
        return sum([integrand_val[k]*(x[k+1]-x[k]) for k in range(n-1)])  
    
    def J1_eval(self, x):
        
        '''
        estimate the J1 function of a distribution x
        
        '''
        # sort x
        x.sort()
        n = len(x)
        # compute the integrand of J1
        intgrd_val = [np.sqrt(k/n)*np.sqrt(1-k/n) for k in range(1,n)]
        res = sum([intgrd_val[k]*(x[k+1]-x[k]) for k in range(n-1)])
        
        return res
    
    def add_intercept(self, df):
        
        '''
        add intercept column to the design matrix df
        '''
        m = df.shape[0]
        if len(df.shape) == 1:
            df = df.reshape((m, 1))
        a = np.array([1]*m).reshape((m,1))
        return np.concatenate((a,df), axis=1)
    
    def model_ls(self, A, Y):
        
        '''
        solve least squares AX = Y
        '''
        if len(Y.shape) == 1:
            Y = Y.reshape(len(Y), 1)
        X = np.linalg.lstsq(A, Y, rcond=None)[0]
        fit = A@X
        residuals = Y - fit
        sd = np.std(residuals, axis = 0)
        res = {'Beta': X, 'fit': fit,
               'residuals': residuals, 'sd': sd}
        return res
    
    def integration(self, x, y, f):
        
        '''
        compute the integral of f with respect to the empirical CDF of x union y
        '''
        x = np.array(x).flatten()
        y = np.array(y).flatten()
        x_y = np.sort(np.concatenate((x, y)))
        point_val = np.array([f(x_y[i])*(x_y[i+1]-x_y[i]) for i in range(len(x_y)-1)])
        return np.sum(point_val, axis = 0)    
    
    def cdf_inverse(self, f):
        
        '''
        compute the inverse of a CDF f
        '''
        
        def f_inverse(alpha):
            if f(self._low)>=alpha:
                return -np.inf
            elif f(self._up)<=alpha:
                return np.inf
            else:
                def g(x):
                    return f(x) - alpha
                init_up = 1
                init_low = -1
                while f(init_up)<=alpha:
                    init_up *= 2
                while f(init_low)>=alpha:
                    init_low *= 2
                result = spo.root_scalar(g, bracket=[init_low, init_up])
                return result.root
        return f_inverse
    
    def sort_2d(self, A):
        
        '''
        sort a 2d distribution
        '''
        A = np.array(A)
        if len(A.shape)!=2:
            raise TypeError('Incorrect dimension.')
        else:
            m, n = A.shape
            stop = False
            N = 0
            while stop == False & N<=1000:
                A_temp = A.copy()
                for i in range(m):
                    A[i,:] = np.sort(A[i,:])
                for j in range(m):
                    A[:,j] = np.sort(A[:,j])
                if np.max(A_temp-A)==0:
                    stop = True
                N += 1
        return A
        

class MFDistributionLearning(Utility):
    np.random.seed(123456790)      
    '''
    This class implements the cvMDL algorithm, including both estimation and exploitation; the current
    version can only handle QoI dimension up to 3d
    '''
    
    def __init__(self, weight=None):
        np.random.seed(123456790)
        super().__init__()
        self.alpha = 4  # regularization parameter (large values means no regularization)
        self.q = 0.05 # quantile truncation parameter for estimation in 1d
        self.N = 1000 # 1d exact integral pamameter
        self.NN = 50 # multidimensional integral mesh parameter
        
        # weight function is used in index computation
        if weight == None:    
            def weight_function(x):
                return 1
            self._weight = weight_function
        else:
            self._weight = weight
     
    def index_cvMDL(self, df_high, df_cv, cost):
        
        '''
        compute indices k1 and k2 in cvMDL
        df_high: high-fidelity data
        df_cv: control variates data constructed from low-fidelity models
        cost: cost vector corresponding to the model  
        '''
        # exploitation cost
        c_ept = sum(cost)
        
        # get min/max of empirical CDF of df_high in each coordinate
        l_min = df_high.min(axis=0)
        l_max = df_high.max(axis=0)
        
        # get quantile of df_cv if df_cv is 1d
        if df_cv.shape[1] == 1:
            l_tau_lower = np.quantile(df_cv, self.q)
            l_tau_upper = np.quantile(df_cv, 1-self.q)
        
        # get max of df_high, df_cv columnwise
        max_high_cv = np.zeros(df_high.shape)
        for i in range(df_high.shape[1]):
            max_high_cv[:,i] = np.maximum(df_high[:,i], df_cv[:,i])
        
        # compute empirical CDF for df_high, df_cv, and their max(df_high, df_cv)
        a_Y = MvECDF(df_high)
        a_Z = MvECDF(df_cv)
        a_max = MvECDF(max_high_cv)
        
        # integrand of model selection indices
        def k_1_2_integrand(z):
            if type(z) != list and type(z) != np.ndarray:
                z = np.array([z])
            Y = np.array([1]*df_high.shape[0]) 
            Z = np.array([1]*df_high.shape[0])
            for i in range(df_high.shape[1]):
                Y *= np.less_equal(df_high[:,i], z[i])*1
                Z *= np.less_equal(df_cv[:,i], z[i])*1
            A = self.add_intercept(Z)
            res = self.model_ls(A, Y)
            
            # coompute k1 and k2
            k1 = np.mean(res['sd']**2)*self._weight(z)
            k2 = self._weight(z)*np.std(Y)**2 - k1
            return np.array([k1, k2])
        
        # whole range 
        def model_coef(z):
            t1 = a_Y.ecdf_eval(z)
            t2 = a_Z.ecdf_eval(z)
            t3 = a_max.ecdf_eval(z)
            # tail continuity estimation
            if df_high.shape[1] == 1:
                if t2*(1-t2) != 0:
                    alpha_z = (t3-t1*t2)/(t2*(1-t2))
                elif t2 == 0:
                    t1_l = a_Y.ecdf_eval(l_tau_lower)
                    t2_l = a_Z.ecdf_eval(l_tau_lower)
                    t3_l = a_max.ecdf_eval(l_tau_lower)
                    alpha_z = (t3_l-t1_l*t2_l)/(self.q*(1-self.q))
                else:
                    t1_r = a_Y.ecdf_eval(l_tau_upper)
                    t2_r = a_Z.ecdf_eval(l_tau_upper)
                    t3_r = a_max.ecdf_eval(l_tau_upper)
                    alpha_z = (t3_r-t1_r*t2_r)/(self.q*(1-self.q))
            else:
                if t2*(1-t2) == 0:
                    alpha_z = 0
                else:
                    alpha_z = (t3-t1*t2)/(t2*(1-t2))
            return t1-alpha_z*t2, alpha_z
        
        # partial range
#         def model_coef(z):
            t1 = a_Y.ecdf_eval(z)
            t2 = a_Z.ecdf_eval(z)
            t3 = a_max.ecdf_eval(z)
            # tail continuity estimation
            if df_high.shape[1] == 1:
                if self.q<t2<1-self.q:
                    alpha_z = (t3-t1*t2)/(t2*(1-t2))
                elif t2<= self.q:
                    t1_l = a_Y.ecdf_eval(l_tau_lower)
                    t2_l = a_Z.ecdf_eval(l_tau_lower)
                    t3_l = a_max.ecdf_eval(l_tau_lower)
                    alpha_z = (t3_l-t1_l*t2_l)/(self.q*(1-self.q))
                else:
                    t1_r = a_Y.ecdf_eval(l_tau_upper)
                    t2_r = a_Z.ecdf_eval(l_tau_upper)
                    t3_r = a_max.ecdf_eval(l_tau_upper)
                    alpha_z = (t3_r-t1_r*t2_r)/(self.q*(1-self.q))
            else:
                if t2*(1-t2) == 0:
                    alpha_z = 0
                else:
                    alpha_z = (t3-t1*t2)/(t2*(1-t2))
            return t1-alpha_z*t2, alpha_z
        
        def model_coef_no(z):
            t1 = a_Y.ecdf_eval(z)
            t2 = a_Z.ecdf_eval(z)
            t3 = a_max.ecdf_eval(z)
            # tail continuity estimation
            if t2*(1-t2) == 0:
                alpha_z = 0
            else:
                alpha_z = (t3-t1*t2)/(t2*(1-t2))
            return t1-alpha_z*t2, alpha_z
        
        # compute k1 and k2; only work for 1/2/3D
        if df_high.shape[1] == 1:
            if df_high.shape[0] <= self.N:
                k1, k2 = self.integration(df_high.flatten(), df_cv.flatten(), k_1_2_integrand)
            else:
                k1, k2 = np.mean(np.array([k_1_2_integrand(z)*(l_max[0]-l_min[0]) for z in np.linspace(l_min[0], l_max[0], 1000)]), axis = 0)
        elif df_high.shape[1] == 2:
            L1 = np.linspace(l_min[0], l_max[0], self.NN)
            L2 = np.linspace(l_min[1], l_max[1], self.NN)
            d1, d2 = (l_max - l_min)/self.NN
            k1 = 0
            k2 = 0
            for i in range(self.NN):
                for j in range(self.NN):
                    k1_add, k2_add = k_1_2_integrand([L1[i], L2[j]])
                    k1 += k1_add*d1*d2
                    k2 += k2_add*d1*d2
        elif df_high.shape[1] == 3:
            L1 = np.linspace(l_min[0], l_max[0], self.NN)
            L2 = np.linspace(l_min[1], l_max[1], self.NN)
            L3 = np.linspace(l_min[2], l_max[2], self.NN)
            d1, d2, d3 = (l_max - l_min)/self.NN
            k1 = 0
            k2 = 0
            for i in range(self.NN):
                for j in range(self.NN):
                    for l in range(self.NN):
                        k1_add, k2_add = k_1_2_integrand([L1[i], L2[j], L3[l]])
                        k1 += k1_add*d1*d2
                        k2 += k2_add*d1*d2
        else:
            raise TypeError("Data should be 1d, 2d, or 3d.") 
        res = {'k1': k1, 'k2': c_ept*k2, 'rho': model_coef, 'rho_no': model_coef_no}
        return res        
    
    def cvMDL_pars(self, df_low, df_high, cost, tol_size):    
        '''
        cvMDL parameters
        '''
         # exploration cost
        c_epr = sum(cost)
        
        # get exploration rate m and # low-fidelity models
        m, n_low = df_high.shape[0], len(df_low)
        model_list = self.model_list(n_low, tol_size)
        n_model = len(model_list)
        
        # get least squares for each model
        model_res = []
        model_names = list(df_low.keys())
        for S in model_list:
            df_low_1 = df_low[model_names[S[0]]]
            if len(S)>1:
                for i in range(1,len(S)):
                    df_low_1 = np.concatenate((df_low_1, df_low[model_names[S[i]]]), axis=1)
            model_res.append(self.model_ls(self.add_intercept(df_low_1), df_high))
 
        # get df_cv for each model
        model_cv = [model_res[k]['fit'] for k in range(n_model)]
        
        # get k1/k2 for each model
        k_1_2 = [self.index_cvMDL(df_high = df_high, df_cv = model_cv[k], 
                cost = [cost[i] for i in model_list[k]]) for k in range(n_model)] 
        res = {' '.join([str(list(df_low.keys())[elem]) for elem in model]): (np.sqrt(c_epr*r['k1'])+np.sqrt(r['k2']))**2 \
               for model, r in zip(model_list, k_1_2)}
        return res
    
    def index_aetcd(self, sd, df_high, df_res, model, cost):
        
        '''
        compute indices k1 and k2 in aetcd
        
        '''
        s = len(model)
         
        # exploitation cost
        c_ept = sum(cost)
        
        # compute k1 and k2
        k1 = (2*np.sqrt(s+1)*sd+self.J1_eval(list(df_res)))**2
        k2 = c_ept*self.J1_eval(list(df_high))**2
        
        res = {}
        res['k1'] = k1
        res['k2'] = k2
        
        return res
 
    def model_select_cvMDL(self, df_low, df_high, cost, tol_size, budget):
        
        '''
        select model S for exploitation (cdf)
        df_low: a dictionary
        df_high: a 1/2/3d array with high-fidelity data with fidelity arranged columnwise
        cost: a list of cost parameters associated with the concatenated df_low and df_high 
        tol_size: maximum size of the low-fidelity models for linear regression
        budget: total budget for exploration and exploitation
        '''
        
        # exploration cost
        c_epr = sum(cost)
        
        # get exploration rate m and # low-fidelity models
        m, n_low = df_high.shape[0], len(df_low)
        model_list = self.model_list(n_low, tol_size)
        n_model = len(model_list)
        
        # get least squares for each model
        model_res = []
        model_names = list(df_low.keys())
        for S in model_list:
            df_low_1 = df_low[model_names[S[0]]]
            if len(S)>1:
                for i in range(1,len(S)):
                    df_low_1 = np.concatenate((df_low_1, df_low[model_names[S[i]]]), axis=1)
            model_res.append(self.model_ls(self.add_intercept(df_low_1), df_high))
 
        # get df_cv for each model
        model_cv = [model_res[k]['fit'] for k in range(n_model)]
        
        # get k1/k2 for each model
        k_1_2 = [self.index_cvMDL(df_high = df_high, df_cv = model_cv[k], 
                cost = [cost[i] for i in model_list[k]]) for k in range(n_model)]
        

        # estimated optimal exploration rate
        model_opt = [budget/(c_epr + np.sqrt(c_epr*k['k2']/(k['k1'] + self.alpha**(-m)))) for k in k_1_2]
        # estimated optimal expploration rate at current time
        model_use = [max(m,s) for s in model_opt]
        # estimated optimal loss at current time
        model_risk = [k_1_2[s]['k2']/(budget-c_epr*model_use[s])+(k_1_2[s]['k1']+self.alpha**(-m))/model_use[s] for s in range(n_model)]
        # estimated optimal model
        n_opt = np.argmin(model_risk)

        # bisection exploration
        if model_opt[n_opt]>2*m:
            m_next = 2*m
        elif model_opt[n_opt]>m:
            m_next = int(np.ceil((m + model_opt[n_opt])/2))
        else:
            m_next = m
        
        # optimal exploitation cost
        c_opt = sum(cost[s] for s in model_list[n_opt])
        
        res = {'exp_rate': m_next, 'model': [model_names[s] for s in model_list[n_opt]],
              'coef': model_res[n_opt]['Beta'], 'cost': c_opt, 'sample_rate': int(np.floor((budget-m_next*c_epr)/c_opt)),
              'rho': k_1_2[n_opt]['rho'], 'rho_no': k_1_2[n_opt]['rho_no']}
        return res
        
    def model_select_aetcd(self, df_low, df_high, cost, tol_size, budget):
        
        '''
        select model S for exploitation (cdf)
        df_low: a dictionary
        df_high: a 1d array with high-fidelity data with fidelity arranged columnwise
        cost: a list of cost parameters associated with the concatenated df_low and df_high 
        tol_size: maximum size of the low-fidelity models for linear regression
        budget: total budget for exploration and exploitation
        
        '''
        
        # exploration cost
        c_epr = sum(cost)
        
        # get exploration rate m and # low-fidelity models
        m, n_low = df_high.shape[0], len(df_low)
        model_list = self.model_list(n_low, tol_size)
        n_model = len(model_list)
        
        # get least squares for each model
        model_res = []
        model_names = list(df_low.keys())
        for S in model_list:
            df_low_1 = df_low[model_names[S[0]]]
            if len(S)>1:
                for i in range(1,len(S)):
                    df_low_1 = np.concatenate((df_low_1, df_low[model_names[S[i]]]), axis=1)
            model_res.append(self.model_ls(self.add_intercept(df_low_1), df_high))
        
        # get k1/k2 for each model
        k_1_2 = [self.index_aetcd(sd = model_res[k]['sd']**2, df_high = df_high, df_res = model_res[k]['residuals'], 
                                  model = model_list[k], cost = [cost[i] for i in model_list[k]]) for k in range(n_model)]
        
        # estimated optimal exploration rate
        model_opt = [budget/(c_epr + (c_epr**2*k['k2']/(k['k1']+self.alpha**(-m)))**(1/3)) for k in k_1_2]
        # estimated optimal exploration rate at current time
        model_use = [max(m,s) for s in model_opt]
        # estimated optimal loss at current time
        model_risk = [(k_1_2[s]['k1']/model_use[s])**(1/2)+(k_1_2[s]['k2']/(budget-c_epr*model_use[s]))**(1/2) for s in range(n_model)]
        # estimated optimal model
        n_opt = np.argmin(model_risk)
        S_opt = model_list[n_opt]
        
        # bisection exploration
        if model_opt[n_opt]>2*m:
            m_next = 2*m
        elif model_opt[n_opt]>m:
            m_next = int(np.ceil((m + model_opt[n_opt])/2))
        else:
            m_next = m
        
        # optimal exploitation cost
        c_opt = sum([cost[s] for s in model_list[n_opt]])
        
        res = {'exp_rate': m_next, 'model': [model_names[s] for s in model_list[n_opt]], 'coeff': model_res[n_opt]['Beta'], 'noise_sd': model_res[n_opt]['sd'], \
               'sample_rate': int(np.floor((budget-m_next*c_epr)/c_opt)), 'residuals': model_res[n_opt]['residuals'], 'cost': c_opt
              }
        return res
    
    def cvMDL(self, df_low, df_high, cost, tol_size, budget):
        
        '''
        cvMDL
        '''
        c_epr = sum(cost)
        m = sum(x.shape[1] for x in list(df_low.values())) + 2
        low_keys = list(df_low.keys())
        if m>= budget/c_epr:
            return "The budget is too small."
        else:
            e = 1
            while e>0:
                df_low_temp = {key: df_low[key][0:m,:] for key in low_keys}
                res = self.model_select_cvMDL(df_low = df_low_temp, df_high = df_high[0:m,:], 
                                cost = cost, tol_size = tol_size, budget = budget)
                e = res['exp_rate'] - m
#                 print('Current exploration rate: {}'.format(m),
#                       'Current optimal model: {}'.format(res['model']), 
#                        'More exploration: {}'.format(e>0), sep = ", ")
                m = res['exp_rate']
            res['hf_data'] = df_high
            
            model_res = []
            model_names = list(df_low.keys())
            model = res['model']
            df_low_opt = df_low[model[0]][:m,:]
            if len(model)>1:
                for i in range(1, len(model)):
                    df_low_opt = np.concatenate((df_low_opt, df_low[model[i]][:m,:]), axis=1)
            res['lf_data'] = self.add_intercept(df_low_opt)@res['coef']
            return res

    def aetcd(self, df_low, df_high, cost, tol_size, budget):
        
        '''
        aetc-d algorithm
        
        ''' 
        c_epr = sum(cost)
        m = sum(x.shape[1] for x in list(df_low.values())) + 2
        low_keys = list(df_low.keys())
        if df_high.shape[1]>1:
            return "High-fidelity data needs to be scalar-valued."
        else:
            if m>= budget/c_epr:
                return "The budget is too small."
            else:
                e = 1
                while e>0:
                    df_low_temp = {key: df_low[key][0:m,:] for key in low_keys}
                    res = self.model_select_aetcd(df_low = df_low_temp, df_high = df_high[0:m,:], 
                                    cost = cost, tol_size = tol_size, budget = budget)
                    e = res['exp_rate'] - m
                    m = res['exp_rate']
                return res

    def ecdf_cv(self, df_tt, res, tail=True):
        
        '''
        ecdf estimator based on cvMDL result
        df_tt: a dictionary of exploitation data for estimation
        res: a dictionary given by the cvMDL output
        '''
        model = res['model']
        if tail==True:
            rho = res['rho']
        else:
            rho = res['rho_no']
        df_ttt = df_tt[model[0]][0:res['sample_rate'],:]
        if len(model)>1:
            for i in range(1, len(model)):
                df_ttt = np.concatenate((df_ttt, df_tt[model[i]][0:res['sample_rate'],:]), axis=1)
        df = self.add_intercept(df_ttt)
        emulator = MvECDF(df@res['coef'])
        def cdf_cv(x):
            return max(min(emulator.ecdf_eval(x)*rho(x)[1] + rho(x)[0],1),0)
        return cdf_cv
     
    def ecdf_aetcd(self, df_tt, res, addnoise=True):
        
        '''
        ecdf estimator based on aetcd result
        
        '''
        model = res['model']
        df_ttt = df_tt[model[0]][0:res['sample_rate'],:]
        if len(model)>1:
            for i in range(1, len(model)):
                df_ttt = np.concatenate((df_ttt, df_tt[model[i]][0:res['sample_rate'],:]), axis=1)
        df = self.add_intercept(df_ttt)
        emulator = (df@res['coeff']).flatten()
        if addnoise == False:
            cdf = ECDF(emulator)
        else:
            emulator += np.random.choice((res['residuals']).flatten().tolist(), size=res['sample_rate'], replace=True)
            cdf = ECDF(emulator)
        
        return cdf 
    
    def ecdf_cv_1d(self, df_tt, res):
        
        '''
        exact_sorted cvMDL in 1d
        
        '''
        if res['hf_data'].shape[1] > 1:
            raise TypeError("Output must be 1d.")
        else:
            model = res['model']
            rho = res['rho']
            df_ttt = df_tt[model[0]][0:res['sample_rate'],:]
            if len(model)>1:
                for i in range(1, len(model)):
                    df_ttt = np.concatenate((df_ttt, df_tt[model[i]][0:res['sample_rate'],:]), axis=1)
            df = self.add_intercept(df_ttt)
            temp = df@res['coef']
            emulator = MvECDF(temp)
            def cdf_cv(x):
                return max(min(emulator.ecdf_eval(x)*rho(x)[1] + rho(x)[0],1),0)
            x_discrete = np.sort(np.concatenate([res['hf_data'].flatten(),
                                            res['lf_data'].flatten(),
                                            temp.flatten()]))
            y_discrete = np.sort(np.array([cdf_cv(x) for x in x_discrete]))
            x_min, x_max = min(x_discrete), max(x_discrete)
            cdf_temp = interp1d(x_discrete, y_discrete)
            def cdf(x):
                if x<=x_min:
                    return 0
                elif x>=x_max:
                    return 1
                else:
                    return cdf_temp(x)
            
            return cdf
    
    def cvMDL_exploit_1d(self, df_tt, res):
        
        '''
        cvMDL exploitation output in 1d with both sorting and tail estimates
        
        '''
        cvMDL_est = self.ecdf_cv_1d(df_tt, res)
        cvMDL_quant = self.cdf_inverse(cvMDL_est)
        return {'cdf': cvMDL_est, 'quantile': cvMDL_quant}
    
    def aetcd_exploit_1d(self, df_tt, res):
        
        '''
        aetc-d exploitation output
        
        '''
        aetcd_est = self.ecdf_aetcd(df_tt, res)
        aetcd_quant = self.cdf_inverse(aetcd_est)
        return {'cdf': aetcd_est, 'quantile': aetcd_quant}
```

## Create training and testing data

We have $4000$ samples in total. To ensure the oracle CDF is relatively accurate, we use $3630$ high-fidelity samples to compute the ECDF. The remaining dataset (consisting of at most $370$ samples of both high- and low-fidelity models in total) is used for exploration. Such division guarantees that the oracle CDF uses approximately ten times the samples used by the ECDF computed under the exploration budget. A good choice for the budget is $B = 40000$. To make most of the limited data, for each simulation we randomly select $3630$ samples for exploitation. 

A few things to note:

* The budget is relatively small. (This is a practical problem and requiring more samples for simulation purpose is extremely expensive.) In this case, the model misspecification effects have not appeared yet in the AETC-d algorithm. As a result, AETC-d demonstrates a competitive/superior performance than cvMDL. We may decide later if we want to include AETC-d results also. 

* Since we have already verified the consistency of cvMDL in the previous examples, we may focus more on the accuracy of the estimated CDF under a particular budget. For instance, we may report the overall $L^2$ error ($\int_{\mathbb R}|\widehat{F}(x)-F(x)|^2 dx$), tail statistics such as VaR and CVaR, moment statistics such as mean, variance, skewness, etc. 

* The current use of exploitation samples is not quite right since they are correlated with the samples used to compute the oracle CDF. To ensure independent (low-fidelity) samples for exploitation, we may use the new low-fidelity samples for the limiting model generated later. For the chosen budget, I would expect the limiting model selected by the exploration process is stable. 

Moreover, given the inverse CDF $F^{-1}$, the statistics can be using inverse sampling. 
An implementation of the above simulation is given as follows (a fixed sequence of seed is used to ensure the experiment can be repeated):

## Implementation

```python
import os
# Load data 
df1 = pd.read_csv('exploration_output.csv').loc[1:].to_numpy()[:,1:]
df2 = pd.read_csv('exploitation_output.csv').loc[1:].to_numpy()[:,1:]
df_full = np.vstack((df1, df2))/1e6

from scipy.stats import skew

# results holder
# cvMDL
cvMDL_l2 = [] # overall L^2 error
cvMDL_var = [] # 0.99-VaR   
cvMDL_cvar = [] # 0.99-CVaR
cvMDL_cvar2 = [] # 0.95-CVaR
cvMDL_mean = [] # mean
cvMDL_sd = [] # standard deviation
cvMDL_skewness = [] # skewness
# ecdf
ecdf_l2 = []
ecdf_var = []
ecdf_cvar = []
ecdf_cvar2 = []
ecdf_mean = []
ecdf_sd = []
ecdf_skewness = []
# oracle
oracle_var = []
oracle_cvar = []
oracle_cvar2 = []
oracle_mean = []
oracle_sd = []
oracle_skewness = []

# total number of simulations
N = 50
# maximum number of exploration samples
m = 500
# cvMDL input parameters
cost = [9.1,1.3,1.1,1,108.9]
tol_size = 4
budget = 50*1e3

foldername = f'result_{budget}'
if not os.path.exists(foldername):
    os.makedirs(foldername)
else:
    print("Folder already exists.")

# VaR/CVaR and moment parameters
alpha = 0.99
alpha2 = 0.95
# interval for CVaR computation
I = np.linspace(alpha,1-1e-8,1000)
# interval for moment statistics
I2 = np.linspace(alpha2,1-1e-8,1000)
# interval for moment statistics
II = np.linspace(1e-8,1-1e-8,1000)
# interval for overall error estimation
III = np.linspace(1.9,2.8,1000)
cvMDL_exp = 0

for i in range(N):
    print('i =', i)
    # randomly divide data into exploration and exploitation
    # this needs to be modified once getting new low-fidelity data
    np.random.seed(1234*i)
    S = np.random.permutation(df_full.shape[0])
    df_full = df_full[S,:]
    df_low = {i+1: df_full[:,i:(i+1)][:m,:] for i in range(4)}
    df_high = df_full[:,4:5][:m,:]
    df_tt = {i+1: df_full[:,i:(i+1)][m:,:] for i in range(4)}
    
    # compute oracle CDF
    # this is random due to the data division step given limited data
    y_oracle = ECDF(df_full[:,-1][m:])
    u = Utility()
    np.random.seed(123456790+i)
    y_oracle_inv = u.cdf_inverse(y_oracle)
    
    # cvMDL exploration
    a = MFDistributionLearning()
    cvMDL_res = a.cvMDL(df_low, df_high, cost, tol_size, budget)
    #print('cvMDL selected model:', cvMDL_res['model'])
    print('cvMDL: {}'.format({key: value for key, value in cvMDL_res.items() if key in ['model','exp_rate', 'sample_rate']}))
    cvMDL_exp = cvMDL_exp + cvMDL_res['exp_rate']
    
    # naive ecdf using high-fidelity samples only
    ecdf = ECDF(df_high.flatten()[:int(budget/cost[-1])])
    ecdf_inv = a.cdf_inverse(ecdf)
    
    # cvMDL exploitation
    cvMDL_exploit = a.cvMDL_exploit_1d(df_tt, cvMDL_res)
    cvMDL_cdf = cvMDL_exploit['cdf']
    cvMDL_cdf_inv = cvMDL_exploit['quantile']
    
    # compute mean errors
    cvMDL_l2.append(np.sum((y_oracle(III)-np.array([cvMDL_cdf(x) for x in III]))**2)*(III[1]-III[0]))
    ecdf_l2.append(np.sum((y_oracle(III)-ecdf(III))**2)*(III[1]-III[0]))
    print('L2 error: {}'.format({key: value for key, value in zip(['cvMLD', 'ecdf'], [cvMDL_l2[-1], ecdf_l2[-1]])}))
    
    # compute VaR/CVaR
    oracle_var.append(y_oracle_inv(alpha))
    cvMDL_var.append(cvMDL_cdf_inv(alpha))
    ecdf_var.append(ecdf_inv(alpha))
    print('VaR: {}'.format({key: value for key, value in zip(['oracle','cvMLD', 'ecdf'], [oracle_var[-1], cvMDL_var[-1], ecdf_var[-1]])}))
    
    oracle_cvar.append(np.sum(np.array([y_oracle_inv(x) for x in I])*(I[1]-I[0]))/(1-alpha))
    cvMDL_cvar.append(np.sum(np.array([cvMDL_cdf_inv(x) for x in I])*(I[1]-I[0]))/(1-alpha))
    ecdf_cvar.append(np.sum(np.array([ecdf_inv(x) for x in I])*(I[1]-I[0]))/(1-alpha))
    print('CVaR: {}'.format({key: value for key, value in zip(['oracle','cvMLD', 'ecdf'], [oracle_cvar[-1],cvMDL_cvar[-1], ecdf_cvar[-1]])}))
    
    oracle_cvar2.append(np.sum(np.array([y_oracle_inv(x) for x in I2])*(I2[1]-I2[0]))/(1-alpha2))
    cvMDL_cvar2.append(np.sum(np.array([cvMDL_cdf_inv(x) for x in I2])*(I2[1]-I2[0]))/(1-alpha2))
    ecdf_cvar2.append(np.sum(np.array([ecdf_inv(x) for x in I2])*(I2[1]-I2[0]))/(1-alpha2))
    print('CVaR2: {}'.format({key: value for key, value in zip(['oracle','cvMLD', 'ecdf'], [oracle_cvar2[-1],cvMDL_cvar2[-1], ecdf_cvar2[-1]])}))
    
    # compute moment statistics
    # inverse sampling
    oracle_dt = df_full[:,-1][m:]
    cvMDL_dt = np.array([cvMDL_cdf_inv(x) for x in II])
    ecdf_dt = df_high.flatten()[:int(budget/cost[-1])]
    # mean
    oracle_mean.append(np.mean(oracle_dt))
    cvMDL_mean.append(np.mean(cvMDL_dt))
    ecdf_mean.append(np.mean(ecdf_dt))
    print('mean: {}'.format({key: value for key, value in zip(['oracle','cvMLD', 'ecdf'], [oracle_mean[-1],cvMDL_mean[-1], ecdf_mean[-1]])}))
    # standard deviation
    oracle_sd.append(np.std(oracle_dt))
    cvMDL_sd.append(np.std(cvMDL_dt))
    ecdf_sd.append(np.std(ecdf_dt))
    print('std: {}'.format({key: value for key, value in zip(['oracle','cvMLD', 'ecdf'], [oracle_sd[-1],cvMDL_sd[-1], ecdf_sd[-1]])}))
    # skewness
    oracle_skewness.append(skew(oracle_dt))
    cvMDL_skewness.append(skew(cvMDL_dt))
    ecdf_skewness.append(skew(ecdf_dt))
    print('skewness: {}'.format({key: value for key, value in zip(['oracle','cvMLD', 'ecdf'], [oracle_skewness[-1],cvMDL_skewness[-1], ecdf_skewness[-1]])}))
    print('-------------------------------------------------------------------------------------------------------------')
    print('-------------------------------------------------------------------------------------------------------------')
cvMDL_exp = cvMDL_exp/N
print(cvMDL_exp)
```

```python
import csv

def write_results_to_csv(filename, results):
    with open(filename, 'w', newline='') as csvfile:
        result_writer = csv.writer(csvfile, delimiter=',')
        for result in results:
            result_writer.writerow(result)

filename = f'data_{budget}.csv'
with open(filename, 'w', newline='') as csvfile:
    result_writer = csv.writer(csvfile) 
    result_writer.writerow(['Iteration','cvMDL_l2','ecdf_l2','oracle_mean','cvMDL_mean','ecdf_mean','oracle_SD','cvMDL_SD','ecdf_SD','oracle_cvar0.99','cvMDL_cvar0.99','ecdf_cvar0.99','oracle_cvar0.95','cvMDL_cvar0.95','ecdf_cvar0.95'])
    for i in range(N):
        result_writer.writerow([i+1, cvMDL_l2[i], ecdf_l2[i], oracle_mean[i], cvMDL_mean[i], ecdf_mean[i], oracle_sd[i], cvMDL_sd[i], ecdf_sd[i], oracle_cvar[i], cvMDL_cvar[i], ecdf_cvar[i], oracle_cvar2[i], cvMDL_cvar2[i], ecdf_cvar2[i]])

filename2 = f'results_{budget}.csv'
with open(filename2, 'w', newline='') as csvfile:
# MRD 
    result_writer = csv.writer(csvfile) 
    result_writer.writerow(['avg_cvMDL_l2','avg_ecdf_l2','avg_cvMDL_mean','avg_ecdf_mean','avg_cvMDL_SD','avg_ecdf_SD','avg_cvMDL_cvar0.99','avg_ecdf_cvar0.99','avg_cvMDL_cvar0.95','avg_ecdf_cvar0.95'])
# L2
    avg_cvMDL_l2 = np.mean(cvMDL_l2)
    avg_ecdf_l2 = np.mean(ecdf_l2)
# CVAR0.99 
    abs_diff_cvMDL = np.abs((np.array(cvMDL_cvar) - np.array(oracle_cvar))/np.array(oracle_cvar))
    avg_cvMDL_cvar = np.mean(abs_diff_cvMDL)
    abs_diff_ecdf = np.abs((np.array(ecdf_cvar) - np.array(oracle_cvar))/np.array(oracle_cvar))
    avg_ecdf_cvar = np.mean(abs_diff_ecdf)
# CVAR0.95 
    abs_diff_cvMDL2 = np.abs((np.array(cvMDL_cvar2) - np.array(oracle_cvar2))/np.array(oracle_cvar2))
    avg_cvMDL_cvar2 = np.mean(abs_diff_cvMDL2)
    abs_diff_ecdf2 = np.abs((np.array(ecdf_cvar2) - np.array(oracle_cvar2))/np.array(oracle_cvar2))
    avg_ecdf_cvar2 = np.mean(abs_diff_ecdf2)
# Mean
    abs_diff_cvMDL = np.abs((np.array(cvMDL_mean) - np.array(oracle_mean))/np.array(oracle_mean))
    avg_cvMDL_mean = np.mean(abs_diff_cvMDL)
    abs_diff_ecdf = np.abs((np.array(ecdf_mean) - np.array(oracle_mean))/np.array(oracle_mean))
    avg_ecdf_mean = np.mean(abs_diff_ecdf)
# SD 
    abs_diff_cvMDL = np.abs((np.array(cvMDL_sd) - np.array(oracle_sd))/np.array(oracle_sd))
    avg_cvMDL_sd = np.mean(abs_diff_cvMDL)
    abs_diff_ecdf = np.abs((np.array(ecdf_sd) - np.array(oracle_sd))/np.array(oracle_sd))
    avg_ecdf_sd = np.mean(abs_diff_ecdf)
# 
    result_writer.writerow([avg_cvMDL_l2, avg_ecdf_l2, avg_cvMDL_mean, avg_ecdf_mean, avg_cvMDL_sd, avg_ecdf_sd, avg_cvMDL_cvar, avg_ecdf_cvar, avg_cvMDL_cvar2, avg_ecdf_cvar2]) 
```
